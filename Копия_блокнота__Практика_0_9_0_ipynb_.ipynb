# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def unique(a,b):
    a_1 = []
    b_1 = []
    for i in a:
        if i in  set(a).difference(set(b)):
            a_1.append(i)

    for i in b:
        if i in  set(b).difference(set(a)):
            b_1.append(i)
    return a_1,b_1


a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
result=unique(a,b)
print(f'Уникальные элементы в первом списке:{result[0]}')
print(f'Уникальные элементы во втором списке: {result[1]}')
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
def adoption(name,types,age):
    return f'Поздравляем! Вы усыновили животное: {types} по имени: {name}, его возраст: {age}'


def entertainment(action):
    if action.lower() == 'покормить':
        return 'Теперь ваш питомец сыт'

    elif action.lower() == 'поиграть':
        return 'Теперь ваш питомец доволен :)'

    elif action.lower() == 'уложить спать':
        return 'Теперь ваш пиомец полон сил'

    else:
        return 'Выбрано неправильное дейстиве,проверьте введенные данные'


def state():
    return f'Голод: {hunger}\nРадость: {joy}\nЭнергия:{energy}'


hunger = 10
joy = 10
energy = 10
print('Усыновление питомца')
types = input('Какое животное вы хотите усыновить: ')
name = input('Назовите вашего питомца: ')
age = input('Сколько ему лет?: ')
print(adoption(name,types,age))
while True:
    print("Главное меню:")
    print("1. Покормить питомца")
    print("2. Поиграть с питомцем")
    print("3. Уложить питомца спать")
    print("4. Показать состояние питомца")
    print("5. Сдать питомца в приют")

    choice = int(input("Выберите номер пункта меню: "))
    if choice == 1:
        print(entertainment('покормить'))
        energy -= 1
        hunger += 1
    elif choice == 2:
        print(entertainment('поиграть'))
        hunger -= 1
        joy += 1
    elif choice == 3:
        print(entertainment('уложить спать'))
        energy += 1
        joy -= 1
    elif choice == 4:
        print(state())
    elif choice == 5:
        print('Ваш питомец теперь грустит в приюте(')
        break
    else:
        print('Номер не распознан, проверьте введенные данные')
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import random


def fight(damage_knights,health_knights,damage_dragons,health_dragons):
    while health_dragons > 0 and health_knights > 0:
        health_dragons -= damage_knights
        health_knights -= damage_dragons
        if health_dragons<0:
            return 'Вы победили дракона!'
        elif health_knights<0:
            return 'К сожалению, вы проиграли'

while True:
    knights = ['Ариодант','Артас','Ганелон','Роланд','Фолькер','Гатс']
    armors = ['Стёганые доспехи','Кожаные доспехи','Клёпаная кожа','Кольчуга',
              'Пластинчатые доспехи']
    weapons = ['меч','копьё','лук и стрелы','боевой топор','булава']
    damage_knight = random.randint(1,10)
    health_knight = random.randint(20,40)
    knight = random.choice(knights)
    arms = random.choice(armors)
    weaponry = random.choice(weapons)

    dragons = ['Дрогон','Рейегаль','Визерион',' Аркон','Гидра']
    dragon = random.choice(dragons)
    damage_dragon = random.randint(1,10)
    health_dragon = random.randint(20,40)


    print('Добро пожаловать в мир рыцарей и драконов')
    print('Пришло ваше время стать рыцарем! Ваш персонаж:')
    print(f'Рыцарь: {knight}, доспехи: {arms}, оружие: {weaponry}, '
        f'здоровье: {health_knight}, урон: {damage_knight}')
    print('Ваш противник на сегодня:')
    print(f'Дракон: {dragon}, здоровье: {health_dragon}, урон {damage_dragon}')
    print(f'Отправляйтесь на поиски приключений!')
    print(f'Перейдя через лес, вы выходите на небольшую поляну и видит там одинокого'
        'воина,который потерял своë оружие и очень боится не дожить до следующего'
        'утра. Воину требуется помощь в изготовлении нового оружия.')
    choice = input('Помочь ему?Напишите: да или нет ')
    if choice.lower() == 'да':
        print('Вы помогли воину и прокачали свои навыки!(+урон)')
        damage_knight += 1
    else:
        print('Вы продолжаете свой путь')

    print('После знакомства с воином вы продолжаете свой путь, через некоторое'
        'время вы понимаете,что солнце уже садиться и приближается ночь'
        'Это значит, что настало время разбивать лагерь.')
    choice = input('1. Поохотиться или 2. пойти спать?Ввведите номер выбранного варианта ')
    if choice == '1':
        print('Теперь вы сыты, а так как вы быстро поймали добычу, ещё осталось'
            'время для сна. Теперь вы полны сил для новых приключений(+ здоровье)')
        health_knight += 8
    else:
        print('Вы выспались, но уже долгое время ничего не ели, из за чего вы'
            'не в лучшей форме(-здоровье)')
        health_knight -= 5

    print(' На следующий день вы выходите к окраине небольшого городка и видите '
        'таверну. Вы заходите туда. Хозяин таверны просит вас доставить важные'
        'послания.')
    choice = input('Помочь ему?Напишите: да или нет ')
    if choice.lower() == 'да':
        print('Вы потратили много сил на это задание(-здоровье)')
        health_knight -= 4
    else:
        print('Вы продолжаете свой путь')

    print('Вы подходите к горам,в которых по легендам водятся драконы.'
        'Перед собой вы видите 2 дороги')
    choice = input('Куда пойти 1. направо или 2. налево?'
                'Ввведите номер выбранного варианта' )
    if choice == '1':
        print('Вы проходите через святой источник, испив оттуда воды, '
            'вы чувствуете, как энергия наполняет вас (+ здоровье)')
        health_knight += 7
    else:
        print('Вся тропа густо заросла растениями, вам пришлось срубить некоторые'
            'из них, тем самым отточив свои навыки(+урон)')
        damage_knight += 2
    print('Легенды не врут! На вашем пути встречается дракон. Вы вступаете в схватку....')
    print(fight(damage_knight,health_knight,damage_dragon,health_dragon))

    choice=input(f'1. Начать игру заново \n2. Выйти из игры \nВвведите номер выбранного варианта ')
    if choice == '2':
        break

````
